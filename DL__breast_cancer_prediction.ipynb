```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
#loading the data form sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
```

```python
data_frame['label'] = breast_cancer_dataset.target
```

```python
x = data_frame.drop(columns = 'label',axis =1)
y = data_frame['label']
```

```python
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state=2)
```

```python
#this should be done only after splitting training and testing
#we are doing standardize the data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train_std = scaler.fit_transform(x_train)
x_test_std = scaler.fit_transform(x_test)
```

```python
print(x_train_std)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

# **from now on neural networks starts**

```python
#importing tensorflow and keras, in colab we don't need to install this package using !pip
import tensorflow as tf
tf.random.set_seed(3) #same as random_state
from tensorflow import keras #keras uses tensorflow as a backend
```

```python
#setting up the layers of neural network

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),# flatten because we need 1-D, so we use flatten to pass each features separately.
    keras.layers.Dense(20,activation = 'relu'),
    keras.layers.Dense(20,activation = 'relu'),#dense means all the neurons in the previous layer is connnected to every neurons on this layer. this is first hidden layer
    keras.layers.Dense(2,activation = 'sigmoid') #

])
```

```python
#compiling the neural network
model.compile(
    optimizer = 'adam',
    loss='sparse_categorical_crossentropy', #sparase is used when we working with numericals like 0,1,2.
    metrics=['accuracy']
)
```

```python
#training the neural networks
history =model.fit(x_train_std,y_train,validation_split=0.1,epochs=10)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data','validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training data','validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training data','validation data'], loc = 'upper right')
```

```python
loss, accuracy = model.evaluate(x_test_std, y_test)
```
