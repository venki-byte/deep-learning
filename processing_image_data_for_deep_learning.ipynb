```python
from PIL import Image # pillow library
```

```python
img = Image.open('anime.jpeg')
img_resized = img.resize((200,200))
```

```python
img_resized.save('anime.resized.jpeg')
```

```python
#displaying the image from numpy array

img_res = mpimg.imread('anime.resized.jpeg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

```python
#convert rgb to grayscale using opencv
import cv2
```

```python
img = cv2.imread('anime.jpeg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
print(grayscale_image.shape)
```
